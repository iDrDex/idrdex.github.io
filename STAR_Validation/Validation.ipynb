```python
meta = pd.read_sql_query('select * from balanced_meta;', conn)
names = pd.read_sql_query('select * from analysis', conn)
names = names.set_index('id')
meta['fixed_signed'] = meta.pval_fixed * meta.apply(lambda x: -1 if x['te_fixed'] < 0 else 1, axis=1)
meta['random_signed'] = meta.pval_random * meta.apply(lambda x: -1 if x['te_random'] < 0 else 1, axis=1)

# perform clustering and plot the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
df = meta.groupby(['mygene_entrez']) \
        .filter(lambda x: x.analysis_id.count() == len(names.index)) \
        .set_index(['analysis_id', 'mygene_entrez']) \
        .fixed_signed.unstack()
labels = list(names.analysis_name + " " +
                           names.series_count.astype(str) + " gse " +
                           names.platform_count.astype(str) + " gpl " +
                           names.sample_count.astype(str) + " gsm")
pylab.rcParams['figure.figsize']=10,7
```
