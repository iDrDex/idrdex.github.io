# PROBLEM: Variable phenotype descriptors complicate utility of digital samples

```python
import pandas as pd
from numpy import nan
dengue = pd.read_excel("Dengue_samples.xlsx", "Dengue_samples.GEO.csv")
annotation_summary = dengue.pivot_table(values='GSM',rows=['GSE', 'GPL', 'caseQueries', 'controlQueries', 'sample_class'], cols='phase', aggfunc=len).replace(nan, "")
acuteDengue = dengue[(dengue.sample_class >= 0) \
                     & (dengue.phase.map(lambda x:str(x).endswith("Acute")))]
acute_dengue_summary = acuteDengue.pivot_table(rows = ['GSE'], values = ['GSM'], aggfunc=pd.Series.count, cols=['sample_class'])
```

# Manual curation is tedious!

```python
annotation_summary
```

```python
meta = pd.read_csv('dss_dhf.acute.meta.csv')
meta.columns = [col.replace(".","_") for col in meta.columns]
meta = meta.query('k>2 and pval_random < 0.001')\
    .sort('pval_random').set_index("myGeneSym")
def uniCat(fields, fieldSep="; "):
    "Unique concatenation of "
    return fieldSep.join([str(i) for i in set(fields)])

samr = pd.read_csv('dss_dhf.samr.acutePhase.csv')
samr['absFc'] = abs(samr.log2foldChange)
samrMeta = samr.dropna(subset=['myGeneSym', 'log2foldChange']) \
    .groupby(["myGeneSym", "direction"]) \
    .agg({"gse": pd.Series.nunique,
          "gpl": pd.Series.nunique,
          "subset": uniCat,
          "absFc": pd.Series.max,
          "Gene Name": uniCat,
          "myGeneId": uniCat}) \
    .query('gpl>3') \
    .reset_index() \
    .set_index(["myGeneSym"]) \
    .sort("absFc", ascending=False)
robust = meta.join(samrMeta, how = "inner")['direction k caseDataCount controlDataCount TE_fixed TE_random pval_fixed pval_random absFc gpl gse'.split()]
```

```python
#set up environent
import pandas as pd
import psycopg2

try:
    conn = psycopg2.connect("dbname='star' user='postgres' host='star.cwxuiazqnkkn.us-west-2.rds.amazonaws.com' password='rendez'")
except:
    print "I am unable to connect to the database"
#set up data
names = pd.read_sql_query('select * from analysis', conn)
analysis = pd.read_sql_query('select * from balanced_meta', conn)
#make dist_matrix
df = analysis.groupby(['mygene_entrez'])\
             .filter(lambda x: x.count() == len(names.index)) \
             .set_index(['analysis_id', 'mygene_entrez']) \
             .te_fixed.unstack()
# perform clustering and plot the dendrogram
# import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from fastcluster import linkage
labels = list(names.analysis_name + " " + 
                           names.series_count.astype(str) + " gse " + 
                           names.platform_count.astype(str) + " gpl " + 
                           names.sample_count.astype(str) + " gsm")
```

# But we can do so much more!

```python
import pandas as pd
ageing = pd.read_csv("ageing.csv")
cols = ageing.columns
doc = ageing[cols[0]]
for col in cols[1:]:
    doc += ageing[col].str.lower()
regex = '(^|\|)\s*age\:\s*(\d+)\s*($|\|)' #match 'age: 36' and nothing else
ages = doc.str.extract(regex)[1].dropna().astype(int)
diseases = ['cancer', 'astrocytoma', 'medulloblastoma', 'melanoma',
            'diabetes', 'lupus|sle', 'dengue',
            'autism', 'cystic fibrosis',
            'alzheimer', 'dementia',
            'schizophrenia','bipolar', 
            'leukemia', 'lymphoma']
ageDocs = pd.DataFrame({"all(N=%s)"%len(ages): ages})
for disease in diseases:
    diseaseAges = ages.ix[ageDocs.index][doc.ix[ageDocs.index].str.contains(disease)].dropna()
    ageDocs = ageDocs.join(pd.DataFrame({'%s (N=%s)'%(disease,len(diseaseAges)):diseaseAges}), how='left')
```

```python
#Age distribution across diseases
pl = ageDocs.boxplot(rot=90, return_type='dict')
plt.xlabel('Disease FTS')
plt.suptitle('Age distribution')
tight_layout()
savefig('aging.png')
```
