```python
from IPython.display import HTML
def getIframe(url, height = 400, width = 800):
    return HTML('<iframe src=%s height=%s width=%s></iframe>'%(url, height, width))
def getWikipedia(page_words, height = 400, width = 800):
    page = "_".join(page_words.split()).title()
    url = "http://en.m.wikipedia.org/wiki/%s"%page
    return getIframe(url, height, width)
def getPubmed(pmid, height = 400, width = 800):
    url = "http://www.ncbi.nlm.nih.gov/m/pubmed/%s"%pmid
    return getIframe(url, height, width)
from IPython.display import YouTubeVideo
def getYouTubeVideo(code, height=400, width=800):
    return YouTubeVideo(code, height=height, width=width)
```

# Aging

```python
getIframe('http://stargeo.org/init/analysis/results?analysis_id=14&order=balanced_meta.pval_random')
```

```python
import pandas as pd
ageing = pd.read_csv("ageing.csv")
cols = ageing.columns
doc = ageing[cols[0]]
for col in cols[1:]:
    doc += ageing[col].str.lower()
regex = '(^|\|)\s*age\:\s*(\d+)\s*($|\|)' #match 'age: 36' and nothing else
ages = doc.str.extract(regex)[1].dropna().astype(int)
%pylab inline
%config InlineBackend.figure_format = 'svg'
diseases = ['cancer', 'astrocytoma', 'medulloblastoma', 'melanoma',
            'diabetes', 'lupus|sle', 'dengue',
            'autism', 'cystic fibrosis',
            'alzheimer', 'dementia',
            'schizophrenia','bipolar', 
            'leukemia', 'lymphoma']
ageDocs = pd.DataFrame({"all(N=%s)"%len(ages): ages})
for disease in diseases:
    diseaseAges = ages.ix[ageDocs.index][doc.ix[ageDocs.index].str.contains(disease)].dropna()
    ageDocs = ageDocs.join(pd.DataFrame({'%s (N=%s)'%(disease,len(diseaseAges)):diseaseAges}), how='left')
```

# Age distribution across diseases

```python
pl = ageDocs.boxplot(rot=90, return_type='dict')
```
